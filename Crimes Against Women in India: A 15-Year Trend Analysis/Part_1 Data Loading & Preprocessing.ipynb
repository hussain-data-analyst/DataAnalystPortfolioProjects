## 4. Data Integration: Merging Datasets for Unified Analysis

```python
crimes_2001_15 = pd.concat([df2001_2012,df2013,df2014,df2015])
```

```python
crimes_2001_15.head()
```

```python
crimes_2001_15.info()
```

- All the features and Date are in float type
- 'Date' column contain the 'Year of crime' so, we will covert it to 'int' type
- all the other columns are showing number of crimes in each year we can convert them to 'int' type

```python
x = ['Year','Rape', 'Kidnapping & Abduction', 'Dowry Deaths', 
     'Assault on women with intent to outrage her modesty', 
     'Insult to modesty of Women', 'Cruelty by Husband or his Relatives', 
     'Importation of Girls']

for y in x:
    crimes_2001_15[y] = crimes_2001_15[y].astype('int')
```

```python
crimes_2001_15.info()
```

### Let's check if there are some duplicate or misspelt values in 'State/Ut' Column

```python
crimes_2001_15['State/Ut'].value_counts().to_frame()
```

**Here are some misspelt values in 'State/Ut' Column**
1. 'A&N Islands'         _instead of_   'A & N Islands'
2. 'Delhi UT'            _instead fo_   'Delhi'
3. 'D&N Haveli'          _instead of_   'D & N Haveli'

```python
crimes_2001_15['State/Ut']= crimes_2001_15['State/Ut'].replace('A&N Islands','A & N Islands')
crimes_2001_15['State/Ut']= crimes_2001_15['State/Ut'].replace('Delhi UT','Delhi')
crimes_2001_15['State/Ut']= crimes_2001_15['State/Ut'].replace('D&N Haveli','D & N Haveli')
```

```python
crimes_2001_15['State/Ut'].value_counts().to_frame()
```
